```python
########## fits dosyasından neat çıktısını alt alta tek bir txt dosyasına kaydetme #########

import os
import numpy as np
from astropy.io import fits

# Specify the directory containing FITS files
fits_directory = '/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/sigma4/'

# Specify the output text file
output_file_path = '/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/sigma4/altaltaflux.txt'

# Get a list of all FITS files in the directory
fits_files = [f for f in os.listdir(fits_directory) if f.endswith('.fits')]

# Open the output text file in write mode
with open(output_file_path, 'w') as output_file:
    # Write header with column names
    # output_file.write("File_Name\tHeader_Info\n")

    # Loop through each FITS file
    for fits_file in fits_files:
        fits_file_path = os.path.join(fits_directory, fits_file)

        # Open the FITS file
        with fits.open(fits_file_path) as hdul:
            # Access the data in the FITS file
            primary_hdu = hdul[2]
            data_transposed = np.transpose(primary_hdu.data)

            # Get file name without extension
            file_name = os.path.splitext(fits_file)[0]

            # Get header information as a string
            header_info = str(primary_hdu.header)

            # Write file name and header information to the output file
            # Use a tab character to separate file name and header info
            output_file.write(f"{file_name}\n{data_transposed}\n")
            #output_file.write(f"{data_transposed}\n")

print("Output file generated successfully.")
```

```python
########## fits dosyasından neat çıktısını tek tek txt dosyasına kaydetme #########


import os
import numpy as np
from astropy.io import fits

# Specify the directory containing FITS files
fits_directory = '/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/neat_out/'

# Specify the output directory for text files
output_directory = '/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/neat_flux_out/'

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Get a list of all FITS files in the directory
fits_files = [f for f in os.listdir(fits_directory) if f.endswith('.fits')]

# Process each FITS file
for fits_file in fits_files:
    fits_file_path = os.path.join(fits_directory, fits_file)

    # Open the FITS file
    with fits.open(fits_file_path) as hdul:
        # Access the data in the FITS file
        primary_hdu = hdul[1]  # Assuming the primary HDU
        data_transposed = np.transpose(primary_hdu.data)

        # Get file name without extension
        file_name = os.path.splitext(fits_file)[0]

        # Create the output text file path in the specified directory
        output_file_path = os.path.join(output_directory, f"{file_name}.txt")

        # Open the output text file in write mode
        with open(output_file_path, 'w') as output_file:
            # Get header information as a string
            header_info = str(primary_hdu.header)

            # Write file name and header information to the output file
            # Use a tab character to separate file name and header info
            output_file.write(f"{data_transposed}\n")

print("Output files generated successfully.")
```

```python
######## altalta olan verileri gruplandırarak her birini ayrı ayrı dosya oluşturma #######

import os

# Giriş dosyasının yolu ve adı
input_file_path = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/sigma4/altaltaflux.txt"
# Çıkış dizininin yolu
output_directory = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/sigma4/deneme/"

# Dosyayı oku
with open(input_file_path, 'r') as input_file:
    lines = input_file.readlines()

# Çıkış dizini kontrolü ve oluşturma
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Her 166 satırda bir dosya oluştur ve içine yaz
for i in range(0, len(lines), 166):
    output_file_path = os.path.join(output_directory, f"dosya_{i//166 + 1}.txt")
    with open(output_file_path, 'w') as output_file:
        output_file.writelines(lines[i:i+166])

print("İşlem tamamlandı. Dosyalar belirtilen dizine kaydedildi.")
```

```python
import os

# Specify the directory containing TXT files
txt_directory = '/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/sigma4/deneme/'

# Get a list of all TXT files in the directory
txt_files = [f for f in os.listdir(txt_directory) if f.endswith('.txt')]

# Process each TXT file
for txt_file in txt_files:
    txt_file_path = os.path.join(txt_directory, txt_file)

    # Open the TXT file and read its content
    with open(txt_file_path, 'r') as file:
        content = file.read()

    # Remove square brackets from the content
    content_without_brackets = content.replace('[', '').replace(']', '')

    # Open the TXT file in write mode and overwrite its content
    with open(txt_file_path, 'w') as file:
        file.write(content_without_brackets)

print("Square brackets removed successfully.")
```

```python
import pandas as pd
import os

excel_dizin = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/sigma4/neat_out.xlsx"
yeni_klasor = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/sigma4/deneme/"

def kontrol_et(excel_dosyasi):
    # Excel dosyasını oku
    df = pd.read_excel(excel_dosyasi, sheet_name="Sayfa1")  # Sayfa adını belirtin
    
    # Satır sayısını döndür
    satir_sayisi = len(df)
    return satir_sayisi

for dosya_adi in os.listdir(excel_dizin):
    dosya_yolu = os.path.join(excel_dizin, dosya_adi)
    
    try:
        satir_sayisi = kontrol_et(dosya_yolu)
        print(f"{dosya_adi}: {satir_sayisi} satır")
        
        if satir_sayisi != 166:
            print(f"Uyarı: {dosya_adi} dosyasındaki satır sayısı beklenenin dışında!\n")
        
        # Yeni çıktı dosyasını oluştur ve kaydet
        yeni_dosya_yolu = os.path.join(yeni_klasor, f"yeni_{dosya_adi}")
        df.to_excel(yeni_dosya_yolu, index=False)
        print(f"Yeni dosya kaydedildi: {yeni_dosya_yolu}\n")
        
    except Exception as e:
        print(f"Hata: {dosya_adi} dosyasını kontrol ederken bir hata oluştu: {e}\n")
```
