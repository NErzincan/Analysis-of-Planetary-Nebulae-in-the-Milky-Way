```python
import pandas as pd

# Excel dosyasını oku
veriler = pd.read_excel('yeni_dosya.xlsx')

# Birinci sütundaki aynı değerlere sahip hücrelerin 3. satırındaki sayısal değerlerin ortalamasını al
ortalamalar = (
    veriler
    .groupby(veriler.columns[0])
    .apply(lambda group: group.iloc[2:].apply(pd.to_numeric, errors='coerce').mean())
)

# Yeni bir sayfa oluştur
with pd.ExcelWriter('yeni_dosya.xlsx', engine='openpyxl', mode='a') as writer:
    # Eğer 'bitti' adında bir sayfa varsa, onu sil
    if 'bitti' in writer.sheets:
        writer.book.remove(writer.sheets['bitti'])
    
    # Ortalamaları yaz
    ortalamalar.to_excel(writer, sheet_name='bitti', header=True)
```

```python
import pandas as pd

# Excel dosyasını okuma
dosya_yolu = 'Hbeta_akı_olanlar.xlsx'  # Excel dosyasının yolunu ve adını buraya ekleyin
df1 = pd.read_excel(dosya_yolu, sheet_name='Sayfa1')
df2 = pd.read_excel(dosya_yolu, sheet_name='Sayfa2')

# İlk listenin birinci sütununu kullanarak ikinci listeden eşleşen satırları bulma
birlesmis_df = pd.merge(df1, df2, on="Fits_Name", how="left")

# Eşleşme yapılan satırlardaki tüm bilgileri birinci listenin sonrasına eklemek
df1 = pd.concat([df1, birlesmis_df.iloc[:, len(df1.columns):]], axis=1)

# Sonuçları gösterme
print(df1)

df1.to_excel('output.xlsx', index=False)
```

```python
def birlestir_ve_kaydet(birlesik_dosya, yeni_dosya, hedef_dosya):
    # Birleştirilmiş dosyadaki verileri oku
    with open(birlesik_dosya, 'r') as f_birlesik:
        veriler_birlesik = {satir.split()[0]: satir.strip().split()[1:] for satir in f_birlesik.readlines()}

    # Yeni dosyadaki verileri oku
    with open(yeni_dosya, 'r') as f_yeni:
        veriler_yeni = {satir.split()[0]: satir.strip().split()[1:] for satir in f_yeni.readlines()}

    # Birleştirilmiş verileri yazacak dosyayı aç
    with open(hedef_dosya, 'w') as hedef:
        # Ortak anahtarları bul ve birleştirilmiş verileri yaz
        for anahtar in set(veriler_birlesik.keys()) & set(veriler_yeni.keys()):
            birlesik_veri = [anahtar] + veriler_birlesik[anahtar] + veriler_yeni[anahtar]
            hedef.write('\t'.join(birlesik_veri) + '\n')

# Kullanım örneği
birlestir_ve_kaydet('birlesik_dosya.txt', 'Halpha_tümkaynaklar.txt', 'yeni_sonuc_dosyasi.txt')
```

```python
import os

def process_fits_lines_file(file_path):
    # Dosyayı aç ve satırları oku
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Dosya adını al (birinci sütun)
    file_name = os.path.basename(file_path)

    # İstenilen değerleri içeren satırları bul
    results = []
    for line in lines:
        columns = line.strip().split()
        if len(columns) == 5:
            second_column = columns[1]
            third_column = columns[2]

            if second_column == "4861.33":
            # İlk durum: 4861.33 değeri varsa, üçüncü sütunu al
                third_column_value = columns[2]
                results.append((file_name, second_column, third_column_value))
            elif second_column == "6562.77":
            # İkinci durum: 6562.77 değeri varsa, üçüncü sütunu al
                third_column_value = columns[2]
                results.append((file_name, second_column, third_column_value))

    return results

def main():
    # İşlem yapmak istediğiniz dizini belirtin
    directory_path = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/spectra/"

    # İşlem yapmak istediğiniz uzantıyı belirtin
    file_extension = ".fits_lines"

    # Sonuçları depolamak için bir liste oluştur
    final_results = []

    # Belirtilen dizindeki tüm .fits_lines dosyalarını bul
    for file_name in os.listdir(directory_path):
        if file_name.endswith(file_extension):
            file_path = os.path.join(directory_path, file_name)

            # Her dosyayı işle ve sonuçları final listesine ekle
            results = process_fits_lines_file(file_path)
            final_results.extend(results)

    # Sonuçları yazdır veya bir dosyaya kaydet
    output_file_path = "Fitsname_Hb_Ha.txt"
    with open(output_file_path, 'w') as output_file:
        for result in final_results:
            output_file.write(f"{result[0]}\t{result[1]}\t{result[2]}\n")

    print(f"İşlem tamamlandı. Sonuçlar '{output_file_path}' dosyasına kaydedildi.")

if __name__ == "__main__":
    main()
```

```python
import os

def process_and_save_fits_lines_file(file_path):
    # Dosyayı aç ve satırları oku, # ile başlayan satırları filtrele
    with open(file_path, 'r') as file:
        lines = [line for line in file.readlines() if not line.startswith("#")]

    # Dosya adını al (birinci sütun)
    file_name = os.path.basename(file_path)

    # İstenilen değerleri içeren satırları bul
    results = []
    for line in lines:
        columns = line.strip().split()
        if len(columns) == 5:
            second_column = columns[1]
            third_column_value = columns[2]

            if second_column == "4861.33":
                # İlk durum: 4861.33 değeri varsa, üçüncü sütunu al
                results.append((file_name, second_column, third_column_value))
            elif second_column == "6562.77":
                # İkinci durum: 6562.77 değeri varsa, üçüncü sütunu al
                results.append((file_name, second_column, third_column_value))

    # Dosyanın ilk satırını ve sondan iki satırını sil
    lines_to_write = results[1:-2]  # İlk satırı ve sondan iki satırı sil
    # lines_to_write = results  # Eğer tüm satırları kullanmak istiyorsanız bu satırı kullanabilirsiniz

    # Dosyayı aç ve sonuçları dosyaya yaz
    with open(file_path, 'w') as file:
        for line_to_write in lines_to_write:
            file.write(f"{line_to_write[0]}\t{line_to_write[1]}\t{line_to_write[2]}\n")

def main():
    # İşlem yapmak istediğiniz dizini belirtin
    directory_path = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/spectra/"

    # İşlem yapmak istediğiniz uzantıyı belirtin
    file_extension = ".fits_lines"

    # Belirtilen dizindeki tüm .fits_lines dosyalarını bul
    for file_name in os.listdir(directory_path):
        if file_name.endswith(file_extension):
            file_path = os.path.join(directory_path, file_name)

            # Her dosyayı işle ve sonuçları dosyaya kaydet
            process_and_save_fits_lines_file(file_path)

    print(f"İşlem tamamlandı. Dosyalar üzerine kaydedildi.")

if __name__ == "__main__":
    main()
```
