```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# Excel dosyasını oku
dosya_yolu = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/08122023.xlsx"
sayfa_adi = "neat_result"
veri = pd.read_excel(dosya_yolu, sheet_name=sayfa_adi)

# Hangi sütunlarda işlem yapacağınızı belirtin
x_sutunu = input("X ekseninde kullanılacak sütun adını girin: ")
y_sutunu = input("Y ekseninde kullanılacak sütun adını girin: ")

# İşlemi yapın ve yeni sütunları oluşturun
veri['X'] = 12 + np.log10(veri[x_sutunu])
veri['Y'] = 12 + np.log10(veri[y_sutunu])

# Grafik çizimi
fig, ax1 = plt.subplots(figsize=(12, 6))

# Scatter plot
ax1.scatter(veri['X'], veri['Y'])

ax1.set_xlabel('12 + LOG(' + x_sutunu + ')')
ax1.set_ylabel('12 + LOG(' + y_sutunu + ')')

# İkinci y ekseni oluştur
ax2 = ax1.inset_axes([1.05, 0, 0.1, 1], transform=ax1.transAxes)

# Boxplot
boxplot = ax2.boxplot(veri['Y'], vert=True, widths=0.2, patch_artist=True, boxprops=dict(facecolor='lightblue'), showfliers=False)

ax2.set_ylabel('Y Dağılımı')
ax2.set_xticks([])  # İkinci y ekseninde x etiketlerini kaldır

# Ayrılan verileri hata çubukları ile göster
if boxplot['fliers']:
    outliers = boxplot['fliers'][0].get_data()[1]
    ax1.errorbar(np.ones_like(outliers) * (max(veri['X']) + 0.1), outliers, fmt='o', color='red', ecolor='red', capsize=5, label='Ayrılan Veriler')

# Grafiği kaydetme
kayit_klasoru = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/görüntüler/"
# Grafiği kaydetme
kayit_adi = "Tan_Parker_fig11.png"  # İstediğiniz dosya adı
kayit_yolu = os.path.join(kayit_klasoru, kayit_adi)
plt.savefig(kayit_yolu)

# Grafiği gösterme
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Excel dosya yollarını al
dosya_yolu_flux = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/08122023.xlsx"
dosya_yolu_neat = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/08122023.xlsx"

# Sayfa isimlerini al
sayfa_adi_flux = "flux_result"
sayfa_adi_neat = "neat_result"

# Hedef başlık ve sütun başlıklarını al
hedef_baslik_flux = float(input("Hedef başlık (flux): "))
hedef_sutun_basligi_neat = input("Hedef sütun başlığı (neat): ")

# İlk sayfa
veri_flux = pd.read_excel(dosya_yolu_flux, sheet_name=sayfa_adi_flux)

# İkinci sayfa
veri_neat = pd.read_excel(dosya_yolu_neat, sheet_name=sayfa_adi_neat)

# Hedef başlığı içeren satırı bul
hedef_satir_flux = veri_flux[veri_flux.iloc[:, 0] == hedef_baslik_flux]

# İlgili sütunu seç
hedef_sutun_neat = veri_neat[hedef_sutun_basligi_neat]

# Logaritmik dönüşümü uygula
log_sutun_neat = np.log10(hedef_sutun_neat)

# Uzunlukları kontrol et
x_verileri = list(map(str, hedef_satir_flux.iloc[0, 1:]))
if len(x_verileri) != len(log_sutun_neat):
    x_verileri = x_verileri[:len(log_sutun_neat)]  # Uzunluğu eşleştir

# Box plot çizimi
plt.figure(figsize=(10, 6))
plt.boxplot(log_sutun_neat, labels=[x_verileri], vert=False, sym=".")
plt.title(f"{hedef_baslik_flux} Değerine Sahip Satırın {hedef_sutun_basligi_neat} Sütunu (log10) - Box Plot")
plt.xlabel(f"{hedef_sutun_basligi_neat} (log10)")
plt.ylabel("Wavelength")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
import os

# Excel dosyasını oku
dosya_yolu = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/08122023.xlsx"
sayfa_adi = "neat_result"
veri = pd.read_excel(dosya_yolu, sheet_name=sayfa_adi)

# Hangi sütunlarda işlem yapacağınızı belirtin
y_sutunu = input("X ekseninde kullanılacak sütun adını girin: ")
x_sutunu = input("Y ekseninde kullanılacak sütun adını girin: ")

# NaN değerleri çıkar
veri = veri.dropna(subset=[x_sutunu, y_sutunu])

# İşlemi yapın ve yeni sütunları oluşturun
veri['X'] = 12 + np.log10(veri[x_sutunu])
veri['Y'] = 12 + np.log10(veri[y_sutunu])

# Grafik çizimi
plt.figure(figsize=(10, 6))
plt.scatter(veri['X'], veri['Y'])
plt.xlabel('12 + LOG(' + x_sutunu + ')')
plt.ylabel('12 + LOG(' + y_sutunu + ')')
plt.title('Scatter Plot of 12 + LOG Values')

# Regresyon analizi
eğim, kesim, r_degeri, p_value, std_err = linregress(veri['X'], veri['Y'])

# Regresyon çizgisi
plt.plot(veri['X'], eğim * veri['X'] + kesim, color='red', label='Regresyon Çizgisi')

# R2 değerini ve eğimi ekrana yazdırma
print(f'R2 Değeri: {r_degeri**2:.4f}')
print(f'Eğim: {eğim:.4f}')

# Grafiği gösterme
plt.legend()
plt.show()
```

```python
import pandas as pd
from astropy.coordinates import Angle

# Excel dosyasını oku
dosya_yolu = '/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/sigma_5/usno_b1.xlsx'
sayfa_adı = 'Sayfa1'
veri = pd.read_excel(dosya_yolu, sheet_name=sayfa_adı)

# RAJ2000 ve DECJ2000 değerlerini al
ra = veri['RAJ2000']
dec = veri['DECJ2000']

# RAJ2000 ve DECJ2000 değerlerini derece cinsine dönüştür
ra_derece = Angle(ra).degree
dec_derece = Angle(dec).degree

# DRAJ2000 ve DDECJ2000 değerlerini hesapla
dra = ra_derece * 15  # Dereceyi saat açısı birimine dönüştür
ddec = dec_derece

# Sonuçları DataFrame'e ekle
veri['DRAJ2000'] = dra
veri['DDECJ2000'] = ddec

# Yeni bir Excel sayfası oluştur
sayfa2 = pd.ExcelWriter(dosya_yolu, engine='openpyxl', mode='a')
veri.to_excel(sayfa2, sheet_name='sayfa2', index=False)
sayfa2.save()
```
