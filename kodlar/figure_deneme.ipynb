```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# Excel dosyasn oku
dosya_yolu = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/08122023.xlsx"
sayfa_adi = "neat_result"
veri = pd.read_excel(dosya_yolu, sheet_name=sayfa_adi)

# Hangi stunlarda ilem yapacanz belirtin
x_sutunu = input("X ekseninde kullanlacak stun adn girin: ")
y_sutunu = input("Y ekseninde kullanlacak stun adn girin: ")

# lemi yapn ve yeni stunlar oluturun
veri['X'] = 12 + np.log10(veri[x_sutunu])
veri['Y'] = 12 + np.log10(veri[y_sutunu])

# Grafik izimi
fig, ax1 = plt.subplots(figsize=(12, 6))

# Scatter plot
ax1.scatter(veri['X'], veri['Y'])

ax1.set_xlabel('12 + LOG(' + x_sutunu + ')')
ax1.set_ylabel('12 + LOG(' + y_sutunu + ')')

# kinci y ekseni olutur
ax2 = ax1.inset_axes([1.05, 0, 0.1, 1], transform=ax1.transAxes)

# Boxplot
boxplot = ax2.boxplot(veri['Y'], vert=True, widths=0.2, patch_artist=True, boxprops=dict(facecolor='lightblue'), showfliers=False)

ax2.set_ylabel('Y Dalm')
ax2.set_xticks([])  # kinci y ekseninde x etiketlerini kaldr

# Ayrlan verileri hata ubuklar ile gster
if boxplot['fliers']:
    outliers = boxplot['fliers'][0].get_data()[1]
    ax1.errorbar(np.ones_like(outliers) * (max(veri['X']) + 0.1), outliers, fmt='o', color='red', ecolor='red', capsize=5, label='Ayrlan Veriler')

# Grafii kaydetme
kayit_klasoru = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/grntler/"
# Grafii kaydetme
kayit_adi = "Tan_Parker_fig11.png"  # stediiniz dosya ad
kayit_yolu = os.path.join(kayit_klasoru, kayit_adi)
plt.savefig(kayit_yolu)

# Grafii gsterme
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Excel dosya yollarn al
dosya_yolu_flux = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/08122023.xlsx"
dosya_yolu_neat = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/08122023.xlsx"

# Sayfa isimlerini al
sayfa_adi_flux = "flux_result"
sayfa_adi_neat = "neat_result"

# Hedef balk ve stun balklarn al
hedef_baslik_flux = float(input("Hedef balk (flux): "))
hedef_sutun_basligi_neat = input("Hedef stun bal (neat): ")

# lk sayfa
veri_flux = pd.read_excel(dosya_yolu_flux, sheet_name=sayfa_adi_flux)

# kinci sayfa
veri_neat = pd.read_excel(dosya_yolu_neat, sheet_name=sayfa_adi_neat)

# Hedef bal ieren satr bul
hedef_satir_flux = veri_flux[veri_flux.iloc[:, 0] == hedef_baslik_flux]

# lgili stunu se
hedef_sutun_neat = veri_neat[hedef_sutun_basligi_neat]

# Logaritmik dnm uygula
log_sutun_neat = np.log10(hedef_sutun_neat)

# Uzunluklar kontrol et
x_verileri = list(map(str, hedef_satir_flux.iloc[0, 1:]))
if len(x_verileri) != len(log_sutun_neat):
    x_verileri = x_verileri[:len(log_sutun_neat)]  # Uzunluu eletir

# Box plot izimi
plt.figure(figsize=(10, 6))
plt.boxplot(log_sutun_neat, labels=[x_verileri], vert=False, sym=".")
plt.title(f"{hedef_baslik_flux} Deerine Sahip Satrn {hedef_sutun_basligi_neat} Stunu (log10) - Box Plot")
plt.xlabel(f"{hedef_sutun_basligi_neat} (log10)")
plt.ylabel("Wavelength")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
import os

# Excel dosyasn oku
dosya_yolu = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/08122023.xlsx"
sayfa_adi = "neat_result"
veri = pd.read_excel(dosya_yolu, sheet_name=sayfa_adi)

# Hangi stunlarda ilem yapacanz belirtin
y_sutunu = input("X ekseninde kullanlacak stun adn girin: ")
x_sutunu = input("Y ekseninde kullanlacak stun adn girin: ")

# NaN deerleri kar
veri = veri.dropna(subset=[x_sutunu, y_sutunu])

# lemi yapn ve yeni stunlar oluturun
veri['X'] = 12 + np.log10(veri[x_sutunu])
veri['Y'] = 12 + np.log10(veri[y_sutunu])

# Grafik izimi
plt.figure(figsize=(10, 6))
plt.scatter(veri['X'], veri['Y'])
plt.xlabel('12 + LOG(' + x_sutunu + ')')
plt.ylabel('12 + LOG(' + y_sutunu + ')')
plt.title('Scatter Plot of 12 + LOG Values')

# Regresyon analizi
eim, kesim, r_degeri, p_value, std_err = linregress(veri['X'], veri['Y'])

# Regresyon izgisi
plt.plot(veri['X'], eim * veri['X'] + kesim, color='red', label='Regresyon izgisi')

# R2 deerini ve eimi ekrana yazdrma
print(f'R2 Deeri: {r_degeri**2:.4f}')
print(f'Eim: {eim:.4f}')

# Grafii gsterme
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veri setini ykle
dosya_yolu = "/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/08122023.xlsx"
sayfa_adi = "flux_result"
df = pd.read_excel(dosya_yolu, sheet_name=sayfa_adi)

# - iaretli olan deerleri NaN olarak deitir ve saysal formata evir
df = df.applymap(lambda x: pd.to_numeric(x, errors='coerce'))

# Hedef deeri ieren satrlar se
hedef_deger = 5006.84
filtre = df.isin([hedef_deger])
secili_satirlar = df[filtre.any(axis=1)]

# Verileri srala ve en ok tekrar eden blgeyi bul
sorted_values = secili_satirlar.values.flatten()
sorted_values.sort()
most_frequent_value = sorted_values[len(sorted_values) // 2]

# Kutu izimi
plt.boxplot(secili_satirlar.dropna().values.T)

# En ok tekrar eden deeri gsteren izgi
plt.axhline(y=most_frequent_value, color='orange', linestyle='--', linewidth=2, label='En ok Tekrar Edilen Deer')

# Hata ubuklar
plt.errorbar(x=1, y=most_frequent_value, yerr=[most_frequent_value - sorted_values[0], sorted_values[-1] - most_frequent_value],
             fmt='none', color='red', elinewidth=2, capsize=5, label='Hata ubuklar')

# Grafik zerine balk ve etiket ekleyelim
plt.title(f'Tm Stunlarda {hedef_deger} Deeri Bulunan Satrlarn Box Plot')
plt.xlabel('Stunlar')
plt.ylabel('Deerler')

# Legend ekleme
plt.legend()

# Grafikleri gsterelim
plt.show()
```

```python
import pandas as pd
from astropy.coordinates import Angle

# Excel dosyasn oku
dosya_yolu = '/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/sigma_5/usno_b1.xlsx'
sayfa_ad = 'Sayfa1'
veri = pd.read_excel(dosya_yolu, sheet_name=sayfa_ad)

# RAJ2000 ve DECJ2000 deerlerini al
ra = veri['RAJ2000']
dec = veri['DECJ2000']

# RAJ2000 ve DECJ2000 deerlerini derece cinsine dntr
ra_derece = Angle(ra).degree
dec_derece = Angle(dec).degree

# DRAJ2000 ve DDECJ2000 deerlerini hesapla
dra = ra_derece * 15  # Dereceyi saat as birimine dntr
ddec = dec_derece

# Sonular DataFrame'e ekle
veri['DRAJ2000'] = dra
veri['DDECJ2000'] = ddec

# Yeni bir Excel sayfas olutur
sayfa2 = pd.ExcelWriter(dosya_yolu, engine='openpyxl', mode='a')
veri.to_excel(sayfa2, sheet_name='sayfa2', index=False)
sayfa2.save()
```
