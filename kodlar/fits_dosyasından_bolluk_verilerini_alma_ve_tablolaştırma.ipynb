```python
from astropy.io import fits
import pandas as pd
import os

# Bilimsel gsterim iin pandas seeneini ayarla
pd.set_option('display.float_format', '{:.2e}'.format)

# Dizin iindeki tm FITS dosyalarn al
fits_directory = '/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/sigma_5/bilgiislem/neat_out/'
fits_file_paths = [os.path.join(fits_directory, file) for file in sorted(os.listdir(fits_directory)) if file.endswith('.fits')]

txt_output_directory = '/Users/nurullah/Desktop/galaxy_abudance/Uzak_Pc/sigma_5/bilgiislem/'

# Dizin var m kontrol et, yoksa olutur
if not os.path.exists(txt_output_directory):
    os.makedirs(txt_output_directory)

for fits_file_path in fits_file_paths:
    # Open the FITS file
    with fits.open(fits_file_path) as hdul:
        # Access the data and header of the LINES BinTableHDU (hdul[2])
        data = hdul[2].data

        # Convert the FITS data to a pandas DataFrame
        df = pd.DataFrame(data)

        # Semek istediiniz stunlar belirleyin (rnein, 71. satrdan 119. satra kadar olan stunlar aln)
        selected_columns = df.iloc[78:119, :]

        # lemleri uygula
        processed_data = pd.DataFrame()
        processed_data['Column_1'] = selected_columns.iloc[:, 0]
        processed_data['Column_2'] = selected_columns.iloc[:, 1].apply(lambda x: f'{x:.2e}')
        processed_data['Column_3'] = selected_columns.iloc[:, 2].apply(lambda x: f'{x:.2e}')
        processed_data['Column_4'] = selected_columns.iloc[:, 3].apply(lambda x: f'{x:.2e}')
        
        # 6. stundaki ilemi gerekletir
        processed_data['Processed_Column_5'] = (selected_columns.iloc[:, 2] - selected_columns.iloc[:, 1]).apply(lambda x: f'{x:.2e}')
        
        # 4. stundaki deer sfrdan kkse toplama ilemi yap, deilse karma ilemi yap
        processed_data['Processed_Column_6'] = selected_columns.apply(lambda row: f'{row[1] + row[3]:.2e}' if row[3] < 0 else f'{row[1] - row[3]:.2e}', axis=1)

        # Dosya isminin ilk alt karakterini al
        file_name = os.path.basename(fits_file_path)[:7]

        # TXT dosyasna bilimsel formatta yaz
        txt_file_path = os.path.join(txt_output_directory, f'{file_name}_output.txt')
        processed_data.to_csv(txt_file_path, sep='\t', index=False)

print(f"Processed data has been successfully exported to {txt_output_directory}")
```

```python
##### bilimsel formatta txt dosyasnda dzenleme yaptktan sonra latex formatnda bolluklar tablolatrma

import os

# Verilerin bulunduu klasr
klasor_yolu = '/Users/nurullah/Desktop/true_PNe/son_alfa_out/neat_out/abundance_text/'

# Klasrdeki txt dosyalarn al
dosya_listesi = sorted([dosya for dosya in os.listdir(klasor_yolu) if dosya.endswith('.txt')])

# LaTeX dosyasnn yolu
latex_dosya_yolu = '/Users/nurullah/Desktop/true_PNe/son_alfa_out/neat_out/tablo.tex'

# Dosyay a ve LaTeX tablosunu yaz
with open(latex_dosya_yolu, 'w') as latex_dosya:
    # LaTeX tablosu iin balk satr
    latex_dosya.write("\\begin{tabular}{|" + "c|" * (len(dosya_listesi) + 1) + "}\n")
    latex_dosya.write("\\hline\n")
    # Stun balklar
    latex_dosya.write(" & " + " & ".join(dosya_listesi) + " \\\\\n")
    latex_dosya.write("\\hline\n")

    # Verileri yaz
    for satir in range(1, 43):
        latex_dosya.write(str(satir))  # Satr bal
        for dosya in dosya_listesi:
            dosya_yolu = os.path.join(klasor_yolu, dosya)
            with open(dosya_yolu, 'r') as dosya_objesi:
                satirlar = dosya_objesi.readlines()
                if satir < len(satirlar):  # Satr says kontrol
                    degerler = satirlar[satir].split()
                    # LaTeX tablosu iin uygun formatta deerleri ekle
                    if len(degerler[1]) >= 4 and len(degerler[4]) >= 4 and len(degerler[5]) >= 4:
                        latex_dosya.write(" & $({}^{{+{}}}_{{- {}}})\\times 10^{{{}}}$".format(
                            degerler[1][:4], degerler[4][:4], degerler[5][:4], degerler[1][-3:]))
                    else:
                        latex_dosya.write(" & $^{{+0}}_{{-}}$")  # Hata durumu iin varsaylan deer
                else:
                    latex_dosya.write(" & $^{{+0}}_{{-}}$")  # Hata durumu iin varsaylan deer
        latex_dosya.write(" \\\\\n")

    # LaTeX tablosunu kapat
    latex_dosya.write("\\hline\n\\end{tabular}")

print(f"LaTeX dosyas {latex_dosya_yolu} adl dosyaya baaryla oluturuldu.")
```
