```python
from astroquery.gaia import Gaia
from astropy import coordinates as coords
from astropy import units as u

# Kaynaklarn bulunduu txt dosyasnn yolu
file_path = "/Users/nurullah/Desktop/galaxy_abudance/RaDec.txt"

# Kaynaklar oku ve ile
with open(file_path, "r") as file:
    for line in file:
        # Satrdaki Ra ve Dec deerlerini al
        ra_str, dec_str = line.strip().split()

        # Ra ve Dec deerlerini uygun formata dntr
        ra_value = f"{ra_str}d"
        dec_value = f"{dec_str}d"
        
        # Koordinatlar olutur
        coo = coords.SkyCoord(ra=ra_value, dec=dec_value, unit=(u.degree, u.degree))

        # Gaia veritabanna sorgu olutur
        query = f"SELECT * FROM gaiadr2.gaia_source WHERE CONTAINS(POINT('ICRS', {coo.ra.deg}, {coo.dec.deg}), circle('ICRS', ra, dec, 0.00083))=1"

        # Sorguyu gnder ve sonular al
        job = Gaia.launch_job(query)
        results = job.get_results()

        # Sonular ile (rnein, yazdr)
        print(f"Ra: {ra_value}, Dec: {dec_value}")
        print(results)
```
